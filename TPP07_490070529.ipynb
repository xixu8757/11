# Data cleaning steps
##### 1. Convert missing value to 0
##### 2. Remove outliers
##### 3. Group sorting

###### Import csv data

```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
df = pandas.read_csv("Marks.csv") #read data
df #show data
```

###### Convert null values to 0

```python
df = df.fillna(0) # convert nulls
df
```

###### M3 null with number 4 is converted to 0
###### Remove outliers from the dataset
*  According to the mark description, mark should range from 0 to 100

```python
h = df[df['M2'] <= 100] # remove outlier from the range 0-100
h
```

###### Data with number 0 is removed
###### Then Group the dataset

```python
h.groupby('SID') # group data
h
```

###### Null values of the data converted to 0, outlier is removed, and the data is grouped
